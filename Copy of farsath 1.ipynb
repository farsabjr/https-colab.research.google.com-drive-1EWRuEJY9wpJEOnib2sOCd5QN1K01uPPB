3. Write a Python program to count the number of strings from a given list of strings. The string length is 2 or more and the first and last characters are the same.

```
Sample List : ['abc', 'xyz', 'aba', '1221']
Expected Result : 2
```

````python
List=['abc','xyz','aba','1221']
def count_strings_with_same_ends(strings):
    count=0
    for word in strings:
        if len(word) >= 2 and word[0]== word[-1]:
           count += 1
    return count

result = count_strings_with_same_ends(List)
print("Number of strings with the same first and last character:",result)

````

17. Write a Python program to get the top three items in a shop.

```
Sample data: {'item1': 45.50, 'item2':35, 'item3': 41.30, 'item4':55, 'item5': 24}

Expected Output:
item4 55
item1 45.5
item3 41.3
```


````python
data={'item1': 45.50,'item2':35,'item3':41.30, 'item4':55,'item5':24}
values=data.values()
values=list(values)
values.sort()
a=values[-3:]
a.reverse()
for k,v in data.items():
  if v in a:
    print(k,v)
````

item3 41.3
item4 55
